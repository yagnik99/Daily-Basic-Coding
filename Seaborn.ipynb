```python
# Import Seaborn
import seaborn as sns
```

##Import Seaborn datasets

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

#**Distribution Plots**

```python
sns.distplot(crash_df['not_distracted']) # Can doe kde= False to remove line, also can use bins to bucket data.
```

##Comparing 2 distributions

```python
sns.jointplot(x = 'speeding', y= 'alcohol', data = crash_df, kind = 'hex')  # kind = reg if want to have regression line between. Can also use 'kde' for KDE plot or hex for hexagonal
```

##**Pair Plot**

```python
sns.pairplot(crash_df)
```

###Loading other dataset for more examples.

```python
tips_df = sns.load_dataset('tips')
tips_df.head(3)
```

```python
sns.pairplot(tips_df, hue='sex')  #hue used to get for categorical data and palete for color
```

#**Styling**

```python
import matplotlib.pyplot as plt
```

```python
sns.set_style('dark')   # To set dark grid, can also set to white, whitegrid, dark and ticks
plt.figure(figsize = (8,4)) # To change fig size

sns.set_context('talk', font_scale=1.4)  # To set context and text, can have paper, talk, poster.

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

#**Categorical Plots**

###**Bar Plots**

```python
sns.barplot(x='sex', y='total_bill', data=tips_df)
```

###**Count Plot**

###To find count of categorical variables in data

```python
sns.countplot(x='sex', data=tips_df)
```

###**Box Plot**

### Shows quartile of data.

```python
sns.set_style('white')   # To set dark grid, can also set to white, whitegrid, dark and ticks
plt.figure(figsize = (10,8)) # To change fig size
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.boxplot(tips_df['total_bill'])
```

###**Violin Plot**

###Similar to box plot.It is combination of box and KDE plot.

```python
plt.figure(figsize = (12,10)) # To change fig size
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

###**Strip Plot**

###Gives scatter points for categorical.

```python
plt.figure(figsize = (10,8)) # To change fig size
sns.stripplot(x='day', y='total_bill', data=tips_df)
```

###Can seperate it further by hue.

```python
plt.figure(figsize = (10,8)) # To change fig size
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

###Can seperatre further for both male and female by using dodge = True.

```python
plt.figure(figsize = (10,8)) # To change fig size
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', dodge=True)
```

#**Palettes**

###Changing colors of data points.

```python
plt.figure(figsize = (10,8)) # To change fig size
sns.set_style('white')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='afmhot',dodge=True) # can use various palette for color change

plt.legend(loc=0)   # Move or set the label. here male/female - or can use other options like - upper right = 1, upper left = 2, lower left = 3, lower right = 4
```
